4. Given a dataframe, sort it by multiple columns

```python
import numpy as np
import pandas as pd
 
df = pd.DataFrame({'Name': ['Raj', 'Rahul', 'Sam', 'Priya'],
                   'Age': [19, 18, 20, 21],
                   'Score': [96, 78, 88, 90]})
 
print('DATAFRAME')
df

df.sort_values(['Age', 'Score'], ascending=[True, False])
```

5. Given a dataframe, select a particular column and display it

```python
import pandas as pd
import numpy as np

exam_data  = {'Name': ['Raj', 'Rahul', 'Sam', 'Priya'],
              'Age': [19, 18, 20, 21],
              'Score': [96, 78, 88, 90]}
labels = ['1', '2', '3', '4']

df = pd.DataFrame(exam_data , index=labels)
print("Select specific columns:")
print(df[['Name', 'Score']])
```

6. Given a dataframe, select rows based on a condition

```python
import pandas as pd
import numpy as np

exam_data  = {'Name': ['Raj', 'Rahul', 'Sam', 'Priya'],
              'Age': [19, 18, 20, 21],
              'Score': [96, 78, 88, 90]}
labels = ['1', '2', '3', '4']

df = pd.DataFrame(exam_data , index=labels)
print(df)

result = df[df['Score'] >=90]  
print(result)
```

11. Write a Pandas program to count the occurrence of a specified substring in a DataFrame column.

```python
import pandas as pd
df = pd.DataFrame({
    'name_code': ['A1','A2','A3', 'A5', 'A6'],
    'date_of_birth ': ['11/09/2003','01/06/1998','15/09/1996','21/02/2022','16/09/1997'],
    'age': [18, 21, 22, 22, 23]
})
print("Original DataFrame:")
print(df)
print("\nCount occurrence of 6 in date_of_birth column:")
df['count'] = list(map(lambda x: x.count("6"), df['date_of_birth ']))
print(df)
```
