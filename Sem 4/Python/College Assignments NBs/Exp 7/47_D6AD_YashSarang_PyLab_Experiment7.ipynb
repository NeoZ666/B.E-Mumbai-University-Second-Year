```python
#CREATING A DATABASE (Pirates.db) AND PERFORMING CRUD OPERATIONS ON IT
import sqlite3
conn = sqlite3.connect('Pirates.db')
conn.execute('''CREATE TABLE Pirates
                (ID INT PRIMARY KEY NOT NULL,
                  NAME TEXT NOT NULL,
                  AGE INT NOT NULL,
                  ADDRESS CHAR(50));''')
conn.close()
```

```python
import sqlite3
conn = sqlite3.connect('Pirates.db')
print("Opened database successfully");
conn.execute("INSERT INTO Pirates (ID,NAME,AGE,ADDRESS) \
              VALUES (1, 'Jack', 21, 'Tortuga')");
conn.execute("INSERT INTO Pirates (ID,NAME,AGE,ADDRESS) \
              VALUES (2, 'Elizabeth', 19, 'Port Royal')");
conn.execute("INSERT INTO Pirates (ID,NAME,AGE,ADDRESS) \
              VALUES (3, 'Will', 20, 'Isla Cruces')");
conn.execute("INSERT INTO Pirates (ID,NAME,AGE,ADDRESS) \
              VALUES (4, 'Davy Jones', 19, 'Davey Jones Locker')");
conn.commit()
print("Records created successfully");
conn.close()
```

```python
import sqlite3
conn = sqlite3.connect('Pirates.db')
cursor = conn.execute("SELECT id, name, address from Pirates")
for row in cursor:
  print("ID = ",row[0])
  print("NAME = ",row[1])
  print("ADDRESS = ",row[2])
```

```python
conn.execute("UPDATE Pirates set ADDRESS = 'The Flying Dutchman' where ID=4")
cursor = conn.execute("SELECT id, name, address from Pirates")
for row in cursor:
  print("ID = ",row[0])
  print("NAME = ",row[1])
  print("ADDRESS = ",row[2])
```
