## 1. Design a Login window and check the credentials of the user. If successful, open up a new window. Otherwise, allow the user to re-enter the credentials (at most 3 attempts).


```python
from tkinter import ttk
from tkinter import *
from tkinter import messagebox
from turtle import width


i = 2


def create_popup(self):
    """ Display pop-up based on password. """
    Id_Pass = [('Hehe', 'idk'), ('YeahBwoi', 'kal aana'),
               ('me nahi bataunga', 'NoobMaster')]

    Id, Password = self.id_ent.get(), self.pw_ent.get()
    global i

    if (Id, Password) in Id_Pass:
        messagebox.showinfo(title='Successful Login',
                            message='Now close')

    elif i >= 0:

        status = messagebox.askyesno(
            title='Try Again?', message='Wrong Credentials,' + str(i) + 'tries remaining\nWould you like to try again?')
        if status == 'yes':
            i -= 1
            root = Tk()
            root.geometry("400x150")
            root.title("Login Page")
            app = Application(root)
            root.mainloop()

        elif status == 'no':
            root.destroy()

    else:
        messagebox.showerror(
            title='Thats It!', message='Maximum tries reached')
        message = "The ID you've entered is invalid."
        root.destroy()


class Application(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        self.grid()
        self.create_widgets()

    def create_widgets(self):
        self.inst_lbl = Label(self, text="Enter Login ID and password")
        self.inst_lbl.grid(row=0, column=0, columnspan=2, sticky=W)

        # create label for ID
        self.id_lbl = Label(self, text="ID: ")
        self.id_lbl.grid(row=1, column=0, sticky=W, columnspan=2)

        # create entry widget to accept ID
        self.id_ent = Entry(self)
        self.id_ent.grid(row=1, column=1, sticky=W, columnspan=2)

        # create label for password
        self.pw_lbl = Label(self, text="Password: ")
        self.pw_lbl.grid(row=2, column=0, sticky=W)

        # create entry widget to accept password
        self.pw_ent = Entry(self, show="*")
        self.pw_ent.grid(row=2, column=1, sticky=W)

        self.submit_bttn = Button(
            self, text='Log In', width=20, height=3, command=lambda: create_popup(self))
        self.submit_bttn.grid(row=5, column=1, ipady=10, ipadx=10)


root = Tk()
root.geometry("400x150")
root.title("Login Page")
app = Application(root)
root.mainloop()
```
