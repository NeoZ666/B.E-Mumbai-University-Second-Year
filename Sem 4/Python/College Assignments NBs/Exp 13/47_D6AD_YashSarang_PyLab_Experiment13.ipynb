# Tasks to be performed 

1. Write a Python code to send a email

```python
import smtplib

gmail_user = 'testh79j@gmail.com'
gmail_password = '' #password is removed after sending the mail

sent_from = gmail_user
to = ['2020.yash.sarang@ves.ac.in']
subject = 'Sending mail through Python using smtplib'
body = 'We are sending mail through smtplib'

email_text = """\
From: %s
To: %s
Subject: %s

%s
""" % (sent_from, ", ".join(to), subject, body)

try:
    smtp_server = smtplib.SMTP_SSL('smtp.gmail.com', 465)
    smtp_server.ehlo()
    smtp_server.login(gmail_user, gmail_password)
    smtp_server.sendmail(sent_from, to, email_text)
    smtp_server.close()
    print ("Email sent successfully!")
except Exception as ex:
    print ("Something went wrong….",ex)
```

2. Write a Python code to read the content of URL

```python
import urllib.request
content = urllib.request.urlopen("https://www.programiz.com/python-programming/examples/hello-world")  
for line in content:  
    print (line) 
```
